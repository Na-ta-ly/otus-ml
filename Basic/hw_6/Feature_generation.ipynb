```python
data_base = pd.read_csv('AB_NYC_2019.csv', index_col=0)
```

```python
data_2 = data[['Latitude', 'Longitude']]
data_base_2 = data_base
```

```python
def crime_count(latitude, longitude, base_Latitude, base_Longitude):
    n = (np.sqrt(((base_Latitude-latitude)*85.395)**2
             + ((base_Longitude-longitude)*111.085)**2) < 5).sum()
    return n
```

```python
frames = [data_base_2, data_2]
result = pd.concat(frames, axis=0, ignore_index=True)
result.head()
```

```python
base_Latitude = result['Latitude'].to_list()
base_Longitude = result['Longitude'].to_list()
data_new = result[['latitude', 'longitude']][result['neighbourhood_group']!=np.nan].apply(
    lambda result: crime_count(result['latitude'],
                                  result['longitude'], base_Latitude, base_Longitude), axis=1)
```
