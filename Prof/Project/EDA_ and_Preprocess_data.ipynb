## Оценим распределение данных по признаку 'Overall'

```python
plt.figure(figsize=(5,3))
sns.histplot(data=data['Overall']);
plt.show()
```

```python
data.groupby('Overall').nunique()
```

```python
def show_full_texts(index, data=data, columns=['Question', 'Essay']):
    for column in columns:
        print('-'*5, '\n', column, sep='')
        print(data[column][index])
```

```python
data[data['Overall'] == 1]
```

```python
show_full_texts(918)
```

Видно, что аномальное значение 'Overall' не ошибка, текст эссе не соответствует заданию.

```python
data[data['Overall'] == 3]
```

```python
show_full_texts(124)
```

```python
show_full_texts(249)
```

Наблюдаем неравномерное распределение. Это необходимо будет учесть при разделении данных на train и test.
Кроме того, классы с 'Overall' 1.0 и 3.0 слишком малочисленны и они не являются ошибками, поэтому их отбросим.

```python
data['Overall'] = data['Overall'].map(str)
data['Task_Type'] = data['Task_Type'].map(str)
```

# Лемматизация

```python
lemmatized_data = data.copy().drop(['Question', 'Essay'], axis=1)
```

Лемматизируем с помощью spacy

```python
from prepoc_func import lemmatize_column
```

```python
%%time
lemmatized_data['Question'] = lemmatize_column(data, 'Question')
```

```python
%%time
lemmatized_data['Essay'] = lemmatize_column(data, 'Essay')
```

```python
lemmatized_data = lemmatized_data.drop(lemmatized_data[lemmatized_data['Overall'] == '1.0'].index)
lemmatized_data = lemmatized_data.drop(lemmatized_data[lemmatized_data['Overall'] == '3.0'].index)
```

```python
lemmatized_data = lemmatized_data.drop(lemmatized_data[lemmatized_data['Overall'] == 'nan'].index)
```
