# Разделим выборку на train и test

```python
from sklearn.model_selection import train_test_split
X = data.drop('price_range', axis=1)
y = data['price_range']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

# Проведем шкалирование данных

```python
from sklearn.preprocessing import StandardScaler
from joblib import dump, load

def SS_scaling_initial(data, file_name='SS_scaler.joblib'):
    """Traines scaler, transforms data and saves scaler"""
    scaler = StandardScaler()
    scaler.fit(data)
    dump(scaler, file_name) 
    return pd.DataFrame(scaler.transform(data), columns = data.columns)


def SS_scale(data, file_name='SS_scaler.joblib'):
    """Loads scaler and applies it to data"""
    scaler = load(file_name)
    return pd.DataFrame(scaler.transform(data), columns = data.columns)
```

```python
X_train_sc = SS_scaling_initial(X_train)
```

```python
X_test_sc = SS_scale(X_test)
X_test_sc.head()
```

# Построение моделей

```python
from sklearn.metrics import accuracy_score
```

## GradientBoosting из sklearn

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model = GradientBoostingClassifier().fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
print('Accuracy: ', accuracy_score(prediction, y_test))
```

## CatBoost

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier(logging_level='Silent').fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
print('Accuracy: ', accuracy_score(prediction, y_test))
```

## XGBoost

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
print('Accuracy: ', accuracy_score(prediction, y_test))
```

## LightGBM

```python
from lightgbm import LGBMClassifier
```

```python
model = LGBMClassifier()
model.fit(X_train, y_train)
prediction = model.predict(X_test)
```

```python
print('Accuracy: ', accuracy_score(prediction, y_test))
```

# Пробуем подобрать параметры

```python
results = []
names = []
```

## GradientBoostingClassifier

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[50, 100, 250, 300, 500],
              'max_features':[20, 4, 5],
              'subsample':[0.5, 0.8, 1.0],
              'learning_rate':[0.1, 0.3, 0.05],
              'max_depth':[3, 6, 8, 10]}

seed = 42
```

```python
base_model = GradientBoostingClassifier(random_state=seed)
gbc_model = GridSearchCV(base_model,param_grid=param_grid, verbose=4, n_jobs=6)
gbc_model.fit(X_train, y_train)
print(gbc_model.best_params_)
```

gbc_model.best_params_ {'learning_rate': 0.1, 'max_depth': 8, 'max_features': 20, 'n_estimators': 250, 'subsample': 0.5}

```python
prediction = gbc_model.predict(X_test)
name = 'GradientBoostingClassifier'
results.append(accuracy_score(prediction, y_test))
names.append(name)
print(name)
print('Accuracy: ', accuracy_score(prediction, y_test))
```

## XGBClassifier

```python
param_grid_xgb = {'n_estimators':[50, 100, 250, 300, 500],
              'num_feature':[20, 4, 5],
              'subsample':[0.5, 0.8, 1.0],
              'learning_rate':[0.1, 0.3, 0.05],
              'max_depth':[3, 6, 8, 10]}
#num_features = max_features
```

```python
base_model = XGBClassifier(random_state=seed)
xgbc_model = GridSearchCV(base_model,param_grid=param_grid_xgb, verbose=2, n_jobs=6)
xgbc_model.fit(X_train, y_train)
print(xgbc_model.best_params_)
```

xgbc_model.best_params_ {'learning_rate': 0.1, 'max_depth': 3, 'n_estimators': 250, 'num_feature': 20, 'subsample': 0.5}

```python
prediction = xgbc_model.predict(X_test)
name = 'XGBClassifier'
results.append(accuracy_score(prediction, y_test))
names.append(name)
print(name)
print('Accuracy: ', accuracy_score(prediction, y_test))
```

## LGBMClassifier

```python
param_grid_lgbm = {'n_estimators':[50, 100, 250, 300, 500],
              'colsample_bytree':[1.0, 0.2, 0.25],
              'subsample':[0.5, 0.8, 1.0],
              'learning_rate':[0.1, 0.3, 0.05],
              'max_depth':[3, 6, 8, 10]}
```

```python
base_model = LGBMClassifier(random_state=seed)
lgbmc_model = GridSearchCV(base_model,param_grid=param_grid_lgbm, verbose=2, n_jobs=6)
lgbmc_model.fit(X_train, y_train)
print(lgbmc_model.best_params_)
```

lgbmc_model.best_params_ {'colsample_bytree': 1.0, 'learning_rate': 0.3, 'max_depth': 3, 'n_estimators': 500, 'subsample': 0.5}


```python
prediction = lgbmc_model.predict(X_test)
name = 'LGBMClassifier'
results.append(accuracy_score(prediction, y_test))
names.append(name)
print(name)
print('Accuracy: ', accuracy_score(prediction, y_test))
```

## CatBoostClassifier

```python
param_grid_cbc = {'iterations':[50, 100, 250, 300, 500],
              'rsm':[1.0, 0.2, 0.25],
              'subsample':[0.5, 0.8, 1.0],
              'learning_rate':[0.1, 0.3, 0.05],
              'depth':[3, 6, 8, 10],
              'bootstrap_type':['MVS', 'Bernoulli']}
```

```python
base_model = CatBoostClassifier(logging_level='Silent', random_state=seed)
cbc_model = GridSearchCV(base_model,param_grid=param_grid_cbc, verbose=2, n_jobs=2)
cbc_model.fit(X_train, y_train)
print(cbc_model.best_params_)
```

cbc_model.best_params_ {'bootstrap_type': 'Bernoulli', 'depth': 3, 'iterations': 500, 'learning_rate': 0.1, 'rsm': 1.0, 'subsample': 0.8}

```python
prediction = cbc_model.predict(X_test)
name = 'CatBoostClassifier'
results.append(accuracy_score(prediction, y_test))
names.append(name)
print(name)
print('Accuracy: ', accuracy_score(prediction, y_test))
```

# Подведем результаты

```python
pd.Series(results, index=names)
```

Выводы: подбор параметров улучшил XGBClassifier и LGBMClassifier. Точность GradientBoostingClassifier осталась на прежнем уровне.
Как и ожидалось, эффективность CatBoostClassifier снизилась. Вероятнее всего, это связано с тем, что многие параметры по умолчанию определяются оптимальными вспомогательным алгоритмом. А при поиске GridSearch это значение не попадает в сетку.
Но эта модель даже после настройки параметров вручную также осталась лидером, показав accuracy 0.935.
