```python
import pandas as pd
import numpy as np
import datetime
```

```python
data = pd.read_csv('data.csv')
col_list = ['Manuf_date','Launch_date', 'Protocol_date']
for col_name in col_list:
    data[col_name] = pd.to_datetime(data[col_name], format='%d.%m.%Y',
                                    errors='coerce', dayfirst=True)
data.head()
```

```python
def convertation(manuf_date, income_date, protocol_date):
    '''Returns time spent on analyses
    manuf_date - manufacturing date,
    income_date - date, when the sample come to laboratory,
    protocol_date - date from the protocol'''
    delay = datetime.timedelta(days=1)
    if str(protocol_date) != 'NaT':
        if str(income_date) != 'NaT':
            in_date = income_date
        elif str(manuf_date) != 'NaT':
            in_date = manuf_date + delay
        else:
            return np.nan
        delta = protocol_date - in_date
        if delta.days < -250 or delta.days > 250:
            in_date = in_date.replace(year=protocol_date.year)
            delta = protocol_date - in_date
        if delta.days < 0:
            delta = in_date - protocol_date    
        if delta.days > 250 or delta.days < 0:
            return np.nan
        return int(delta.days)
    return np.nan
```

```python
data['Days_analysis'] = data[col_list].apply(lambda data: \
                convertation(data[col_list[0]], data[col_list[1]], data[col_list[2]]),
                                              axis=1)
```

```python
def sample_count(entry_dates, exit_dates, date):
    '''Returns number of samples, having the date in their range entry_date - exit_date'''
    sample_count = 0
    for i in range(len(exit_dates)):
        start = entry_dates[i]
        finish = exit_dates[i]
        if (date < finish) and (date >= start):
            sample_count += 1
        if finish == start == date:
            sample_count += 1
    return sample_count
```

```python
start_dates = data['Launch_date'].to_list()
finish_dates = data['Protocol_date'].to_list()
data['Samples_number'] = data[['Launch_date']].apply(lambda data: \
                      sample_count(start_dates, finish_dates, data['Launch_date']),\
                      axis=1)
```
